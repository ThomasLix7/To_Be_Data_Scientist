```python
from bs4 import BeautifulSoup as BS
import urllib.request
import re
from IPython.display import HTML
```

```python
r = urllib.request.urlopen('https://www.newscientist.com/article/mg25934573-800-ai-is-evolving-for-its-own-benefit-not-ours/').read()
```

```python
soup=BS(r, "lxml")
type(soup)
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
print(soup.get_text())
```

```python
print(soup.prettify()[0:1000])
```

```python
for link in soup.find_all('a', attrs={'href': re.compile("^http")}):
    print(link)
type(link)
```

```python
file=open("parsed_data.txt","w")
for link in soup.find_all('a', attrs={'href': re.compile("^http")}):
    soup_link = str(link)
    print(soup_link)
    file.write(soup_link)
file.flush()
file.close()
```
